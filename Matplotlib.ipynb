```python
import matplotlib.pyplot as plt

listx = [1,5,7,9,13,16]
listy = [15,50,80,40,70,50]
plt.plot(listx, listy, color="red", lw="5.0", ls="--", label="food")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

listx1 = [1,5,7,9,13,16]
listy1 = [15,50,80,40,70,50]
plt.plot(listx1, listy1, label="Male")
listx2 = [2,6,8,11,14,16]
listy2 = [10,40,30,50,80,60]
plt.plot(listx2, listy2, color="red", linewidth=5, linestyle="--", label="Female")
plt.legend()
plt.xlim(0, 20)
plt.ylim(0, 100)
plt.title("Pocket Money ")
plt.xlabel("Age ")
plt.ylabel("Money ")
plt.show()
```

```python
import matplotlib.pyplot as plt

listx1 = [1,5,7,9,13,16]
listy1 = [15,50,80,40,70,50]
plt.bar(listx1, listy1, label="")
listx2 = [2,6,8,11,14,16]
listy2 = [10,40,30,50,80,60]
plt.bar(listx2, listy2, color="red", label="")
plt.legend()
plt.xlim(0, 20)
plt.ylim(0, 100)
plt.title("")
plt.xlabel("")
plt.ylabel("")
plt.show()
```

```python
import matplotlib.pyplot as plt

labels = ["", "", "", ""]
sizes = [5, 10, 20, 15]
colors = ["red", "green", "blue", "yellow"]
explode = (0, 0, 0.05, 0)
plt.pie(sizes,explode = explode,labels = labels,colors = colors,\
    labeldistance = 1.1,autopct = "%3.1f%%",shadow = True,\
    startangle = 90,pctdistance = 0.6)
plt.axis("equal")
plt.legend()
plt.show()
```
